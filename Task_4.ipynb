## Задача 3
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
p = sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
p.fig.suptitle("Стоимость недвижимости в зависимости от жилой площади")
plt.xlabel('Стоимость недвижимости (в млн.)')
plt.ylabel('Жилая площадь')

 
;
```

### Вывод:  График показывает, что чем больше жилая площадь, тем выше цена недвижимости. 

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5)
plt.xlabel('Стоимость недвижимости (в млн.)')
plt.ylabel('Конструкция здания и дизайн')
plt.title('Зависимость стоимости недвижимости от качества конструкции и дизайна')
;
```

### Вывод: Наибольшую стоимость имеют дома с высоким качеством конструкции и дизайна (значение от 11 до 13).

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Стоимость недвижимости (в млн.)')
plt.ylabel('Оценка вида из окна')
plt.title('Зависимость стоимости недвижимости от вида из окна')
;
```

### Вывод: График показывает, наибольшую стоимость имеют дома с хорошим видом из окна

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='waterfront', y='price', hue='floors')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости (в млн.)')
plt.title('Распределение стоимости недвижимости по виду на набережную и этажам')
plt.legend(title='Количество этажей')
;
```

### Вывод:  График показывает стоимость недвижимости с видом на набережную или без него с учетом количества этажей здания. Дома с видом на набережную и несколькими этажами имеют более высокую стоимость.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Тепловая карта влияния характеристик на стоимость жилья')
;
```

### Вывод: График показывает, что существует линейная зависимость от размера жилой площади и стоимостью недвижимости - чем больше площадь дома, тем выше его стоимость. На стоимость недвижимости также влияет качество конструкции и дизайна, общая площадь, количество ванных комнат в доме.  Дома с видом на набережную и несколькими этажами имеют более высокую стоимость. Дома построенные недавно также имеют более высокую стоимость.
